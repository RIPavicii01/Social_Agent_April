```python
# 1.kbo_league라는 빈 리스트를 만드세요.
kbo_league=[]
# 2.for문을 사용하여 3번 반복하면서 input()으로 팀 이름과 승점을 입력받으세요.
for i in range(3):
    name = input("팀 이름을 입력하세요: ")
    point = int(input(f"{name}의 승점을 입력하세요: "))
    # 3.입력받은 데이터를 {"name": 팀이름, "point": 승점} 형태의 딕셔너리로 만들어 리스트에 append 하세요.
    kbo_league.append({"name": name, "point": point})
# 4.lambda를 활용하거나 정렬 기능을 사용하여 승점(point) 내림차순으로 정렬하세요.
kbo_league.sort(key=lambda x: x["point"], reverse=True)
# 5.최종적으로 "1등: XX팀 (OO점)" 형식으로 출력하세요.
for i in range(len(range(3))):
  print(i+1,"등:",kbo_league[i]["name"])
```

```python
#목표: 한 주간의 평균 식비를 계산하고, 10,000원이 넘는 '과소비' 요일을 찾아내기.
#check_expenses(menu) 함수를 만드세요.
lunch_menu = {"월": 9000, "화": 8500, "수": 12000, "목": 7000, "금": 11000}
#함수 안에서 sum(menu.values())와 len(menu)를 사용해 평균 식비를 구하세요.
def check_expenses(menu):
  total_price = 0
  i = sum(menu.values())
  print(i)
  #for day, price in menu.items():를 사용하여 10,000원이 넘는 요일만 출력하세요.
  for key,value in menu.items():
    if(value>=10000):
      print(key)
    total_price = total_price + value
  #평균 식비를 return 하세요.
  return total_price/len(lunch_menu)
# 아래에 함수를 정의하고 호출해 보세요!



# 함수 호출 및 결과 출력
avg = check_expenses(lunch_menu)
print(f"이번 주 평균 식비는 {avg}원입니다.")
```
