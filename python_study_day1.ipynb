```python
# 엘든 링 데미지 계산 예시
base_damage = 150
buff_multiplier = 1.2
total_damage = base_damage * buff_multiplier

print("최종 데미지:", total_damage)

# 야구 경기 이닝 계산
total_outs = 10
innings = total_outs // 3  # 몇 이닝인가?
remaining_outs = total_outs % 3  # 이닝 종료 후 남은 아웃 카운트는?

print("진행 이닝:", innings)
print("남은 아웃:", remaining_outs)
```

```python
#문제: 배틀그라운드 파밍 계산기
#당신은 현재 배그에서 '5.56mm 탄약' 124발을 가지고 있습니다.

#ammo 변수에 124를 저장하세요.
ammo = 124
#총 한 탄창에는 30발이 들어갑니다. ammo를 활용하여 **"최대로 채울 수 있는 탄창 수(몫)"**와
#**"재장전 후 남는 탄약 수(나머지)"**를 각각 계산하여 출력해 보세요
print(ammo//30)
print(ammo%30)
```

```python
# 오버워치 2 승급 조건 예시
my_win_count = 5
required_wins = 5
is_rank_up = (my_win_count == required_wins)

print("승급 성공?:", is_rank_up)

# 엘든 링 아이템 장착 조건 (근력과 기량이 모두 충족되어야 함)
my_strength = 20
my_dexterity = 15

can_equip = (my_strength >= 18) and (my_dexterity >= 18)
print("무기 장착 가능?:", can_equip)
```

```python
# 문제: 사회복무요원 연가 사용 가능 여부 확인
# 조건: 남은 연가가 0보다 커야 하고, 긴급 상황(is_emergency)이 아니어야(False) 합니다.

# 1. 변수 설정
remaining_leave = 5
is_emergency = False

# 2. 연가 사용 가능 여부 판별 (빈칸을 채워보세요!)
# 힌트: (remaining_leave가 0보다 큰지 비교) and (is_emergency가 False와 같은지 비교)
can_go_home = (remaining_leave > 0) and (is_emergency == False)

# 3. 결과 출력
print("오늘 연가를 쓸 수 있나요?:", can_go_home)
```

```python
# 문제: 엘든 링 아이템 장착 가능 여부 확인
# 조건: 근력(str)이 20 이상 '이면서' 기량(dex)이 15 이상이어야 합니다.

my_str = 22
my_dex = 13

# 아래 변수에 비교 연산자와 논리 연산자(and/or)를 사용하여 조건을 완성하세요.
can_equip = (my_str>=20)and(my_dex>15)
print("무기를 장착하 실수 있습니까?" , can_equip)
```

```python
# 문제: 오버워치 2 나의 모스트 영웅 리스트 관리
# 1. 'heroes' 리스트에 "겐지", "트레이서", "한조"를 순서대로 담으세요.
heroes =["겐지","트레이서","한조"]

# 2. 리스트의 1번 인덱스(두 번째 영웅)를 출력해 보세요.
print(heroes[1])

# 3. 세 번째 영웅인 "한조"를 "카시디"로 수정하고 전체 리스트를 출력해 보세요.
heroes[2] = "카시디"
print(heroes)
```

```python
# 초기 멤버
overwatch_team = ["디바", "트레이서"]

# 신규 영웅 합류 (append)
overwatch_team.append("키리코")
print("영웅 추가 후:", overwatch_team) # ['디바', '트레이서', '키리코']

# 특정 영웅 제외 (remove)
overwatch_team.remove("트레이서")
print("영웅 제외 후:", overwatch_team) # ['디바', '키리코']

# 현재 팀 인원 확인 (len)
print("현재 인원수:", len(overwatch_team)) # 2
```

```python
# 문제: 배틀그라운드 스쿼드 멤버 관리하기

# 1. 'squad' 리스트에 "나", "친구1", "친구2"를 담으세요.
squad = ["나", "친구1", "친구2"]
# 2. 새로운 멤버 "친구3"이 합류했습니다. append()를 사용하여 리스트 끝에 추가하세요.
squad.append("친구3")
# 3. "친구1"이 게임을 종료했습니다. remove()를 사용하여 삭제하세요.
squad.remove("친구1")
# 4. 현재 최종 스쿼드 멤버 리스트와 총 인원수(len)를 출력해 보세요.
print(len(squad))
```
