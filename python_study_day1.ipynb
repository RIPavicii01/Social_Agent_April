```python
# 1.변수 선언 및 데이터 담기
team_name = "LG 트윈스" #문자열(String)
score = 5 #정수(Integer)
batting_average = 0.312 # 실수(Float)
is_game_over = False #불리언(Boolean)

# 2. 데이터 출력하기
print("팀 이름:", team_name)
print("현재 점수:", score)
print("타율:", batting_average)
print("게임 종료 여부:", is_game_over)

# 3. 변수 값 변경 (점수 추가!)
score = score + 2
print("추가 득점 후 점수:", score)
```
